```python
# ğŸ“’ meta_ranker_data_check.ipynb

import json
import pandas as pd
from collections import Counter

# 1. æ–‡ä»¶è·¯å¾„
DATA_PATH = "output/meta_ranker/training_data.json"

# 2. é€è¡Œè¯»å– JSON
records = []
with open(DATA_PATH, "r") as f:
    for line in f:
        try:
            records.append(json.loads(line.strip()))
        except json.JSONDecodeError:
            continue

print(f"âœ… Loaded {len(records)} records")

# 3. è½¬ DataFrameï¼Œæ–¹ä¾¿ç»Ÿè®¡
df = pd.DataFrame(records)

# 4. ç»Ÿè®¡æ­£è´Ÿæ ·æœ¬
label_counts = Counter(df["label"])
print("ğŸ“Š Label distribution:", label_counts)

# 5. æ­£è´Ÿæ¯”ä¾‹
pos_ratio = label_counts[1] / max(1, sum(label_counts.values()))
print(f"âš–ï¸ Positive ratio: {pos_ratio:.4f}")

# 6. æ‰“å°å‡ æ¡æ­£æ ·æœ¬
print("\nğŸ” Example positive samples:")
print(df[df["label"] == 1].head(5).to_dict(orient="records"))
```
